## Manipulate json dictionary

```python
import pandas as pd
```

```python
df = pd.read_json("../data/interim/Gekko_gecko_LN22022_2_UTR.json")
df = df.T
df
```

```python
### switch gene id with name
# iterate over table index
for i in df.index:
    # search for ID and name attribute for gene type features
    attributes = df.loc[i, "attributes"]
    if df.loc[i, "type"] == "gene":
        if 'Name' in attributes.keys():
            # get original attribute value
            original_ID = attributes["ID"]
            original_Name = attributes["Name"]
            # switch value
            df.loc[i, "attributes"]["ID"] = original_Name
            df.loc[i, "attributes"]["Name"] = original_ID
df.loc[[i for i in df.index if 'Name' in df.loc[i, "attributes"].keys()]]
```

## Export modified table to json

```python
df.T.to_json("Gekko_gecko_LN22022_2_UTR_modified.json")
```

## Convert json to gff

```python
%%bash
python ../scripts/funannotate_json_to_gff.py Gekko_gecko_LN22022_2_UTR_modified.json Gekko_gecko_LN22022_2_UTR_modified.gff3
```
